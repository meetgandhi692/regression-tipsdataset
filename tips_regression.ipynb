```python
### Finding the Boundary Values for Outliers
print("For total_bill")
upper_limit = df['total_bill'].mean() + 3*df['total_bill'].std()
lower_limit = df['total_bill'].mean() - 3*df['total_bill'].std()
print("Highest allowed",upper_limit)
print("Lowest allowed",lower_limit)
print(upper_limit,lower_limit)
## Applying the limits
df["total_bill"] = np.where (df["total_bill"] <lower_limit, lower_limit,df['total_bill'])
df["total_bill"] = np.where(df["total_bill"] >upper_limit, upper_limit,df['total_bill'])
print(df.shape)
df.isnull().sum()
```

```python
# Another method 
from sklearn.preprocessing import LabelEncoder
le_sk = LabelEncoder()
label_smoker = le_sk.fit_transform(df['smoker'])
# label_smoker

le_day = LabelEncoder()
label_day = le_day.fit_transform(df['day'])
# label_day

le_time = LabelEncoder()
label_time = le_time.fit_transform(df['time'])
# label_time

le_sex = LabelEncoder()
label_sex = le_sex.fit_transform(df['sex'])
# label_sex

df.index
```

```python
new_df = pd.DataFrame(list(zip(label_smoker.tolist(), label_day.tolist(), label_time.tolist(), label_sex.tolist())),
               columns =['smoker', 'day', 'time', 'sex'])

y1 = df['total_bill']
# Max-Min Normalization
y1_new = (y1-np.min(y1))/(max(y1)-min(y1))
new_df['total_bill']= y1_new

y1 = df['size']
# Max-Min Normalization
y1_new = (y1-np.min(y1))/(max(y1)-min(y1))
new_df['size']= y1_new

print(df['total_bill'].isnull().sum(), df['size'].isnull().sum())
print(new_df['total_bill'].isnull().sum(), new_df['size'].isnull().sum())

new_df
print(new_df.isnull().sum)
# max(y1)-min(y1)

new_df['total_bill']= df['total_bill']
new_df['size']= df['size']
print(new_df.isnull().sum)
```

```python
## Split the data into train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (new_df, Y_data , random_state= 10, test_size=0.2)   
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
print(X.head(5))
print(X_train.head(5))
print(X_test.head(5))
# print(y_test)
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X_train,y_train)
```

```python
print(X_train.columns)
reg.coef_
```

```python
y_pred = reg.predict(np.array(X_test))
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression  ### Using correlation statitics
from sklearn.linear_model import LinearRegression

# configure to select all features
fs = SelectKBest(score_func=f_regression, k='all')
# learn relationship from training data
fs.fit(X_train, y_train)
# transform train input data
X_train_fs = fs.transform(X_train)
# transform test input data
X_test_fs = fs.transform(X_test)

# what are scores for the features
for i in range(len(fs.scores_)):
    print('Feature %d: %f' % (i, fs.scores_[i]))
# plot the scores
plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
plt.show()

# Using selected features
model = LinearRegression()
model.fit(X_train_fs, y_train)
# evaluate the model
y_pred = model.predict(X_test_fs)
# evaluate predictions
mae = mean_absolute_error(y_test, y_pred)
print('MAE: %.3f' % mae)

print('mean_squared_error: %.3f' % mean_squared_error(y_test, y_pred))
```
